```python
# Checking the min , Max, & Average of temeperature  
print ("Minimum temperature : ", df.temperature.min())
print ("Average temperature : ", df.temperature.mean())
print ("maximum temperature : ", df.temperature.max())

#Checking humidity
print ("Minimum humidity : ", df.humidity.min())
print ("Average humidity : ", df.humidity.mean())
print ("maximum humidity : ", df.humidity.max())

#Checking Rainfall

print ("Minimum rainfall : ", df.rainfall.min())
print ("Average rainfall : ", df.rainfall.mean())
print ("maximum rainfall : ", df.rainfall.max())
```

#### Upon checking we can see that the maximum temperature is = 43.67 Dergree ' f ' and the minumum is = 8.82, minimum temperature recorded is = 8.82 ###
 ***based on this observation lets find out the crops that require high temperature and low***

```python
print ("Summer Crops")
print(df[(df['temperature'] >30 ) & (df['humidity'] > 50)]['label'].unique())

print ("Winter crops")
print(df[(df['temperature'] < 20 ) & (df['humidity'] < 35 )]['label'].unique())

print ("Rainy Crops")
print(df[(df['rainfall'] > 200) & (df['humidity'] > 30)]['label'].unique())
```
